```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
import re
from nltk.corpus import stopwords
import spacy
nlp=spacy.load('en_core_web_sm')
from sklearn.feature_extraction.text import TfidfVectorizer
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel("P597 DATASET.xlsx")
```

```python
df.head()
```

```python
df["review"]=df["title"]+ " "+ df["body"]
```

```python
df["review"].head()
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)      # remove links
    text = re.sub(r"[^a-z\s]", "", text)     # remove numbers & symbols
    text = re.sub(r"\s+", " ", text)         # remove extra spaces
    return text

df["clean_review"] = df["review"].apply(clean_text)
```

```python
from textblob import TextBlob

def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0.05:
        return "Positive"
    elif polarity < 0.05:
        return "Negative"
    else:
        return "Neutral"

df["sentiment"] = df["clean_review"].apply(get_sentiment)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report
```

```python
tfidf=TfidfVectorizer()
x=tfidf.fit_transform(df["review"])
```

```python
x.toarray()
```

```python
x.shape
```

```python
svc=SVC(kernel="rbf",gamma=0.3)
svc.fit(x,df["sentiment"])
y_pred=svc.predict(x)
accuracy_score(df["sentiment"],y_pred)
```

```python
print(classification_report(df["sentiment"],y_pred))
```
